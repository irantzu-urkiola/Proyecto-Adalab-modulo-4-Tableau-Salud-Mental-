```python
# EDA b√°sico para entender nuestros DATOS

def eda_basico(df):

    print('üå∑Ejemplo de datos del DF:')
    display(df.head(3))
    display(df.tail(3))
    display(df.sample(3))
    print('________________________________________________________________________________________________________')

    print('üåªN√∫mero de Filas:')
    display(df.shape[0])
    print('________________________________________________________________________________________________________')

    print('üå±N√∫mero de Columnas:')
    display(df.shape[1])
    print('________________________________________________________________________________________________________')

    print('üåºInformaci√≥n de la tabla:')
    display(df.info())
    print('________________________________________________________________________________________________________')

    print('üåëNombre de las columnas:')
    display(df.columns)
    print('________________________________________________________________________________________________________')

    print('üçÑDescripci√≥n de los datos num√©ricos:')
    display(df.describe().T)
    print('________________________________________________________________________________________________________')

    print('üåãDescripci√≥n de los datos no-num√©ricos:')
    display(df.describe(include='object').T)
    print('________________________________________________________________________________________________________')

    print('üçÇSaber si hay datos √∫nicos:')
    display(df.nunique())
    print('________________________________________________________________________________________________________')

    print('üêñQue datos son nulos por columnas:')
    display(df.isnull().sum())
    print('________________________________________________________________________________________________________')

    print('üê≤Filas duplicadas:')
    total_duplicados = df.duplicated().sum()
    if total_duplicados > 0:
        print(f'cantidad de duplicados: {total_duplicados}')
        print('Primeros duplicados')
        display(df[df.duplicated()].head(3))
    else:
        print('No hay duplicados')
    print('________________________________________________________________________________________________________')
```

## Eliminamos columna "code" 

Tiene nulos y no nos aporta al estudio ya que ya disponemos de la columna pa√≠ses. No necesitamos su c√≥digo

```python
archivo = '1- mental-illnesses-prevalence.csv'
nuevo_archivo = '1-mental-illnesses-prevalence_sin_code.csv'

print(f'\nüóÇÔ∏è An√°lisis del archivo: {archivo}')
try:
    df = pd.read_csv(archivo)
    
    # Eliminar la columna 'Code' si existe
    if 'Code' in df.columns:
        df = df.drop(columns='Code')
        print("üßπ Columna 'Code' eliminada correctamente.")

except Exception as e:
    print(f"‚ö†Ô∏è Error al procesar {archivo}: {e}")
```

```python
print("üìÅ Archivo: 1-mental-illnesses-prevalence_sin_code.csv")
print("üßæ Columnas:")
for col in df.columns:
    print(f" - {col}")
```

```python
# Renombrar columnas
df = df.rename(columns={
    'Entity': 'Pa√≠s',
    'Year': 'A√±o',
    'Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized': 'Esquizofrenia',
    'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized': 'Depresi√≥n',
    'Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized': 'Ansiedad',
    'Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized': 'Trastorno Bipolar',
    'Eating disorders (share of population) - Sex: Both - Age: Age-standardized': 'Des√≥rdenes Alimentarios'
})
```

## Redondeamos las columnas float a 2 decimales

```python
# Seleccionar las columnas num√©ricas a redondear
columnas_float = ['Esquizofrenia', 'Depresi√≥n', 'Ansiedad', 'Trastorno Bipolar', 'Des√≥rdenes Alimentarios']

# Redondear a 2 decimales
df[columnas_float] = df[columnas_float].round(2)
```

## Revisamos columna paises

```python
df['Pa√≠s'].unique()
```

```python
# Lista de agrupaciones a separar
agrupaciones_list = [
    'Africa (IHME GBD)',
    'America (IHME GBD)',
    'Asia (IHME GBD)',
    'Europe (IHME GBD)',
    'European Union (27)',
    'High-income countries',
    'Low-income countries',
    'Lower-middle-income countries',
    'Upper-middle-income countries',
    'World'
]

# Extraer pa√≠ses: todos los que NO est√°n en agrupaciones_list
paises = df[~df['Pa√≠s'].isin(agrupaciones_list)]

# Extraer continentes (primer grupo)
continentes = df[df['Pa√≠s'].isin([
    'Africa (IHME GBD)',
    'America (IHME GBD)',
    'Asia (IHME GBD)',
    'Europe (IHME GBD)',
    'European Union (27)'
])]

# Extraer agrupaciones econ√≥micas (segundo grupo)
agrupaciones_economicas = df[df['Pa√≠s'].isin([
    'High-income countries',
    'Low-income countries',
    'Lower-middle-income countries',
    'Upper-middle-income countries'
])]

# Extraer fila 'World'
world = df[df['Pa√≠s'] == 'World']

# Guardar archivos CSV
paises.to_csv('datos_paises.csv', index=False)
continentes.to_csv('datos_continentes.csv', index=False)
agrupaciones_economicas.to_csv('datos_agrupaciones_economicas.csv', index=False)
world.to_csv('datos_world.csv', index=False)
```
